```python
files = os.listdir(os.getcwd())
i = 0
while(True):
    i += 1
    fn = "output" + str(i) + ".avi"
    if fn in files:
        continue
    else:
        out = cv2.VideoWriter(fn,-1,20.0,(480,640))    
        break
        
cam  =  cv2.VideoCapture(0)

a = time.time()
while(cam.isOpened()):
    
    try:
        ret,frame = cam.read()
        out.write(frame)

        if (time.time() - a) > 3:
            break
    except:
        print 'excepted'
        break
        
cam.release()
```
